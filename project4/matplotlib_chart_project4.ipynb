```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack', 'Did you seek any specialist for a treatment?'])
df.plot.box(grid='True')
plt.title('Mental problem of students')
```

```python
depr = df[df['Do you have Depression?']==1]

depr.groupby(['Age'])['Do you have Depression?'].count().plot.bar()
plt.title('repartition of depression depending on age')
```

```python
data = df.groupby(['Age'])['Age'].sum()
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897','#380282','#FF4500','#E50000']

plot = data.plot.pie(y='mass', figsize=(5, 5),wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors);


plt.title("Age repartition of the participants")  ;plt.show()
```
