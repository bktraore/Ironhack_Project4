```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\kyrie\ironhack\project4\cleansheet2.csv")
```

```python
sns.displot(df, x="Age")
plt.title('age repartition')
```

```python
ax = sns.violinplot(df['What is your CGPA?'])
ax
plt.title('violon plot of the grade of the students')
```

```python
sns.relplot(x='Age', y='What is your CGPA?', data=df)
plt.ylabel('What is your CGPA?')
plt.xlabel('Age')
plt.title('scatter plot of the grades of the students depending on their age')
```

```python
sns.set_style("darkgrid")


number =50

x1 = df['Age']
y1 = df['What is your CGPA?']
z1 = df['Do you have Depression?']

plt.figure(figsize=(8,7), dpi = 100)
axes = plt.axes(projection='3d')
print(type(axes))
axes.scatter3D(x1, y1, z1)

axes.set_xlabel('age')
axes.set_ylabel('grade')
axes.set_zlabel('depression')
plt.title('3D plot about the risk of depression depending on age and grade')
plt.show()
```
